```python
   
from __future__ import print_function, division

# Create a class named Point

class Point:
    """Represents a point in 2-D space.

   # attributes: x, y
    """


def print_point(p):
    """Print a Point object in human-readable format."""
    print('(%g, %g)' % (p.x, p.y))


class Rectangle:
    """Represents a rectangle. 

    attributes: width, height, corner.
    """


def find_center(rect):
    """Returns a Point at the center of a Rectangle.
    Functions can return instances. 
    find_center() takes a Rectangle as an argument and 
    returns a Point that contains the coordinates of the 
    center of the Rectangle:


    rect: Rectangle

    returns: new Point
    """
    p = Point()
    p.x = rect.corner.x + rect.width/2.0
    p.y = rect.corner.y + rect.height/2.0
    return p


def grow_rectangle(rect, dwidth, dheight):
    """Modifies the Rectangle by adding to its width and height.
    Objects are mutable. You can change the state of an object 
    by making an assignment to one of its attributes. 
    To change the size of a rectangle without changing its position, 
    you can modify the values of width and height
    
    You can also write functions that modify objects.
    For example, grow_rectangle takes a Rectangle object and two numbers, 
    dwidth and dheight, and adds the numbers to the width and height 
    of the rectangle:

    rect: Rectangle object.
    dwidth: change in width (can be negative).
    dheight: change in height (can be negative).
    """
    rect.width += dwidth
    rect.height += dheight


def main():
    blank = Point()#call the class point. Return value is reference to point object
    blank.x = 3 #assign values to an instance using dot notation
    blank.y = 4
    print('blank', end=' ')
    print_point(blank)

    box = Rectangle()#instantaitate rectangle object

    box.width = 100.0#assign values #to the object

    box.height = 200.0
    box.corner = Point()#instantiate
    box.corner.x = 0.0
    box.corner.y = 0.0

    """Go to the object box refers to 
    and select the attribute named corner;
    then go to that object and select the attribute named x.(fig15.2)""""

    
    center = find_center(box)
    print('center', end=' ')
    print_point(center)

    print(box.width)
    print(box.height)
    print('grow')
    grow_rectangle(box, 50, 100)
    print(box.width)
    print(box.height)


if __name__ == '__main__':
    main()
```
